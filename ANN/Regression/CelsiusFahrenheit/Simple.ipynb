```
try:
  %tensorflow_version 2.x
except Exception:
  pass

import tensorflow as tf
from tensorflow import keras

print(tf.__version__)
```

```
import numpy as np
np.random.seed(7)
```

```
celsius_values    = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
fahrenheit_values = np.array([-40,  14, 32, 46, 59, 72, 100],  dtype=float)

for i,c in enumerate(celsius_values):
  print("{} degrees Celsius = {} degrees Fahrenheit".format(c, fahrenheit_values[i]))
```

```
from keras.layers import Dense
from keras.models import Sequential
```

```
model = Sequential(name='model')
model.add(Dense(units=1, input_shape=[1], name='dense_layer'))
model.summary()
```

```
from keras.optimizers import Adam

learning_rate = 0.2
optimizer = Adam(lr=learning_rate)
```

```
model.compile(loss='mean_squared_error', optimizer=optimizer)
```

```
epochs = 500
```

```
history = model.fit(celsius_values, 
                    fahrenheit_values, 
                    epochs=epochs, 
                    verbose=True)
```

```
import matplotlib.pyplot as plot

plot.xlabel('Epoch number')
plot.ylabel("Mean squared error")
plot.plot(history.history['loss'])
```

```
print(model.predict([100.0]))
```

```
print("These are the layer variables: {}".format(model.get_layer(index=-1).get_weights()))
```
