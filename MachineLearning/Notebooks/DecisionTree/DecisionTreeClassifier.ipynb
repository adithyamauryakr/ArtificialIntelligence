# Decision Tree Classifier

In this example, we will learn to classify a dataset using a [Decision Tree Classifier](https://en.wikipedia.org/wiki/Decision_tree_learning).

We will use an iris dataset for classification.

See [link](https://archive.ics.uci.edu/ml/datasets/iris) for more details.

# Import required python modules.

```python
import numpy as np
np.random.seed(7)
```

```python
import pandas as pd
```

```python
from sklearn import datasets
```

# Load an iris dataset.

### Define feature names.

```python
feature_names = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
```

### Load an iris dataset from sklearn datasets.

```python
iris_dataset = datasets.load_iris()
```

### Define input variables (X) and an output variable (y).

```python
X = pd.DataFrame(iris_dataset.data, columns = feature_names)
y = iris_dataset.target
```

# Import required python modules.

```python
from sklearn.tree import DecisionTreeClassifier
```

# Create a decision tree based classfier.

```python
model = DecisionTreeClassifier()
model.fit(X,y)
```

# Interpret the generated decision tree. 

### Import required python modules.

```python
from sklearn.tree import export_graphviz
```

### Export the generated decision tree as DOT file.

```python
export_graphviz(model, 'iris_tree.dot', feature_names = feature_names)
```

### Export the generated DOT file as PNG image file.

```python
!dot -Tpng iris_tree.dot -o iris_tree.png
```

### Import required python modules.

```python
import matplotlib.pyplot as plt
import cv2
```

### Visualize the generated decision tree.

```python
%matplotlib inline
img = cv2.imread('iris_tree.png')
plt.figure(figsize = (20, 20))
plt.imshow(img)
```
